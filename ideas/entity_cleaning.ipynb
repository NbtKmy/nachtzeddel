````python
import requests
import json
import codecs
import string
import time
import pandas as pd
import os
from tqdm import tqdm
````

````python
# loop to extract places from json files
df = pd.DataFrame()
files = os.listdir('entities')
for file in tqdm(files):
    # print (file)
    # Name of the image file
    myimage = file.replace('.json', '.txt')
    date_folder = ''
    # Parse Json
    with open('entities/' + file, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
        if 'date_folder' in json_data:
            date_folder = json_data['date_folder']
        if 'annotations' in json_data:
            annotations = json_data['annotations']
            for annotation in annotations:
                start = ''
                end = ''
                spot = ''
                confidence = ''
                annotationid = ''
                title = ''
                uri = ''
                label = ''
                categories = ''
                types = ''
                alternateLabels = ''
                place = 0
                person = 0
                if 'start' in annotation :
                    start = annotation['start']
                if 'end' in annotation :
                    end = annotation['end']
                if 'spot' in annotation :
                    spot = annotation['spot']
                if 'confidence' in annotation :
                    confidence = annotation['confidence']
                if 'id' in annotation :
                    annotationid = annotation['id']
                if 'title' in annotation :
                    title = annotation['title']
                if 'uri' in annotation :
                    uri = annotation['uri']
                if 'label' in annotation :
                    label = annotation['label']
                if 'categories' in annotation :
                    categories = annotation['categories']
                if 'types' in annotation :
                    types = annotation['types']
                    if 'http://dbpedia.org/ontology/Place' in types:
                        place = 1
                    if 'http://dbpedia.org/ontology/Person' in types:
                        person = 1
                        # TODO: extract dates from categories, ex. "Geboren 1899","Gestorben 1962"
                if 'alternateLabels' in annotation :
                    alternateLabels = annotation['alternateLabels']
                df = pd.concat([df, pd.DataFrame.from_records([{'image' : myimage,
                                                                'date_folder' : date_folder,
                                                                'start' : start,
                                                                'end' : end,
                                                                'spot' : spot,
                                                                'confidence' : confidence,
                                                                'annotationid' : annotationid,
                                                                'title' : title,
                                                                'uri' : uri,
                                                                'label' : label,
                                                                'categories' : categories,
                                                                'types' : types,
                                                                'alternateLabels' : alternateLabels,
                                                                'place' : place,
                                                                'person' : person
                                                               }])], ignore_index=True)            
````
