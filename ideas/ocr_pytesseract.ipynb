```python
import time
import pandas as pd
from PIL import Image,ImageEnhance
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:/Users/iriarte/AppData/Local/Programs/Tesseract-OCR/tesseract'
import os
# Params
madate = '20240906'
```

```python
print(pytesseract.get_languages(config=''))
```

## convert image to txt

```python
# loop to OCR from files
df = pd.DataFrame()
images = os.listdir('github/nachtzeddel/images')
for image in images :
    print(image)
    text = pytesseract.image_to_string(Image.open('github/nachtzeddel/images/' + image), lang='deu')
    df = df.append({'image' : image, 'text' : text}, ignore_index=True)
    # Name of the file with the results
    myfileout = 'github/nachtzeddel/ideas/ocr_mit_pytesseract/' + image.replace('.jpg', '.txt')

    # create file
    f = open(myfileout, mode='w', encoding='utf-8')
    f.write(text)
    f.close()
```
