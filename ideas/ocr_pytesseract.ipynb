```python
import time
import pandas as pd
from PIL import Image,ImageEnhance
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract'
import os
# Params
madate = '20240906'
```

```python
print(pytesseract.get_languages(config=''))
```

## convert image to txt

```python
# loop to OCR from files
df = pd.DataFrame()
folders_images = os.listdir('images')
folders_results = os.listdir('results')
for folder in folders_images:
    if (folder not in folders_results):
        os.mkdir('results/' + folder)
```

```python
folders_images = os.listdir('images')
for folder in folders_images:
    images = os.listdir('images/' + folder)
    for image in images :
        print(folder + '/' + image)
        text = pytesseract.image_to_string(Image.open('images/' + folder + '/' + image), lang='deu')
        df = pd.concat([df, pd.DataFrame.from_records([{'image' : image, 'text' : text}])], ignore_index=True)
        # Name of the file with the results
        myfileout = 'results/' + folder + '/' + image.replace('.jpg', '.txt')
        # create file
        f = open(myfileout, mode='w', encoding='utf-8')
        f.write(text)
        f.close()
```
